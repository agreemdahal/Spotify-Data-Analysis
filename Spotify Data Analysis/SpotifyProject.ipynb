```python
#TOP 5 POPULAR ARTIST
top_five_artists =df.groupby("artist").count().sort_values(by="song_title", ascending=False)["song_title"][:5]
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

```python
#TOP FIVE LOUDEST TRACKSONG
top_five_loudest_tracks =df[["loudness", "song_title"]].sort_values(by="loudness", ascending=True)[:5]
top_five_loudest_tracks
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y="song_title", data=top_five_loudest_tracks)
plt.title("Top 5 Loudest Tracks")
plt.show()
```

```python
#ARTIST WITH MOST DANCEABLILITY SONG
top_five_artist_danceable_songs = df[["danceability", "song_title", "artist"]].sort_values(by= "danceability", ascending=False) [:5]
top_five_artist_danceable_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability", y="artist", data=top_five_artist_danceable_songs)
plt.title("Top 5 Danceable")
plt.show()
```

```python
#TOP TEN INSTRUMENTAL SONG
top_ten_instrumental_songs = df[["instrumentalness", "song_title", "artist"]].sort_values(by= "instrumentalness", ascending=False) [:10]
top_ten_instrumental_songs
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness", data= top_ten_instrumental_songs, autopct='%1.2f%%', labels=top_ten_instrumental_songs.song_title)
plt.title("Top 10 Instrurmental Songs")
plt.show()
```

```python
#MUSIC FEATURE PLOT
interest_feature_cols =["tempo", "loudness", "acousticness", "danceability", "duration_ms", "energy"
                      , "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature_col in interest_feature_cols:
    pos_data = df[df["target"] == 1][feature_col]
    neg_data = df[df["target"] == 0][feature_col]
    
    plt.figure(figsize=(12, 7))

    sns.histplot(pos_data, bins=30, label="Positive", color="green", kde=True, element="step")
    sns.histplot(neg_data, bins=30, label="Negative", color="red", kde=True, element="step")
    
    plt.legend(loc="upper right")
    plt.title(f"Positive and Negative Histogram Plot for {feature_col}")
    plt.xlabel(feature_col)
    plt.ylabel("Frequency")
    plt.show()
```
